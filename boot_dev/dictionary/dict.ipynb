You don't need to create a dictionary with values already inside. It is common to create a blank dictionary then populate it later using dynamic values. The syntax is the same as getting data out of a key, just use the assignment operator (=) to give that key a value.

```python
planets = {}
planets['Earth']=True #True is the value
planets['Pluto']=False
print(planets)
```

```python
full_names = ["jack bronson", "jill mcarty", "john denver"]
ns = full_names.split()
print(ns)
names_dict = {}
for full_name in full_names:
    # .split() returns a list of strings
    # where each string is a single word from the original
    name_parts = full_name.split()

    # here we update the dictionary
    names_dict[name_parts[0]] = name_parts[1]

print(names_dict)
# Prints: {'jack': 'bronson', 'jill': 'mcarty', 'john': 'denver'}
```

```python
# A list containing three full-name strings.
full_names = [
    "jack bronson",
    "jill mcarty",
    "john denver"
]

# Create an empty dictionary.
# We will store first names as keys
# and last names as values.
names_dict = {}

# Loop through each full-name string in the list.
for full_name in full_names:

    # full_name is one string during each loop iteration.
    #
    # First iteration:
    # full_name = "jack bronson"
    #
    # .split() separates the string at the space
    # and returns a list of strings.
    name_parts = full_name.split()

    # First iteration:
    # name_parts = ["jack", "bronson"]
    #
    # Index 0 is the first name.
    # Index 1 is the last name.
    first_name = name_parts[0]
    last_name = name_parts[1]

    # Add a new key-value pair to the dictionary.
    #
    # First iteration:
    # names_dict["jack"] = "bronson"
    names_dict[first_name] = last_name

# Print the completed dictionary.
print(names_dict)

# Output:
# {'jack': 'bronson', 'jill': 'mcarty', 'john': 'denver'}
```

```python
def get_most_common_enemy(enemies_dict):
    # Handle the empty case FIRST, before any loop logic.
    # If there are no enemies at all, there's nothing to compare,
    # so we return Python's actual None value (not the string "None").
    if not enemies_dict:
        return None

    # These two variables track the "best so far" as we scan the dictionary.
    # We start them as None so the very first enemy we see automatically
    # becomes the current leader (see the comparison logic below).
    best_name = None
    best_count = None

    # .items() gives us BOTH the key (enemy name) and value (count)
    # on each loop pass, instead of just the key like `for enemy in enemies_dict`
    # would. This avoids a second lookup like enemies_dict[enemy] every time.
    for name, count in enemies_dict.items():

        # This condition does two jobs at once:
        # 1. `best_count is None`  -> true only on the very first iteration,
        #    so the first enemy always becomes the initial leader.
        # 2. `count > best_count`  -> STRICTLY greater than (not >=).
        #    This is what makes ties resolve to the FIRST enemy found:
        #    if a later enemy ties the current best, count > best_count
        #    is False, so we don't overwrite the leader.
        if best_count is None or count > best_count:
            best_name = name
            best_count = count

    # After the loop finishes, best_name holds the enemy with the
    # highest count (or the first one seen, in the event of a tie).
    return best_name
```
