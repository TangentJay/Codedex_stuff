# Filtering Profanity from Chat Messages

**Assignment:** Complete `filter_messages(messages)` so that it takes a list of chat message strings and returns **two** new lists:
1. The messages with every instance of the word `"dang"` removed.
2. A parallel list of how many `"dang"`s were removed from each message.

This notebook builds the solution up one concept at a time. Run each cell in order — don't skip ahead, since later cells depend on ideas from earlier ones.

## 1. New tool: `.split()`

`.split()` breaks a string into a list of words, splitting on whitespace by default.

**Why we need it:** a message arrives as one long string. You can't compare "is this word `dang`" against a whole sentence — you need individual words to compare one at a time.

````python
# .split() turns a sentence into a list of words
message = "dang it bobby!"
words = message.split()
print(words)          # ['dang', 'it', 'bobby!']
print(type(words))    # <class 'list'> -- confirm it's actually a list now, not a string
````

## 2. New tool: `.join()`

`.join()` is the reverse of `.split()` — it glues a list of strings back into one string.

Important detail: you call `.join()` **on the separator string**, and pass it the list. Read `" ".join(some_list)` as "join `some_list` together, using a space between each piece."

````python
# .join() glues a list of words back into one sentence
good_words = ["it", "bobby!"]
filtered_message = " ".join(good_words)
print(filtered_message)   # "it bobby!"

# common mistake: calling join on the list instead of the separator
# good_words.join(" ")   # <-- this would raise an AttributeError, lists don't have .join()
````

## 3. Core structure: building two lists in parallel

We need two results per message: the filtered string, and the count. Both get built inside the **same** loop pass, so they stay lined up by index — index `0` of both lists always describes the same original message.

````python
messages = ["dang it bobby!", "look at it go"]

filtered_messages = []   # will hold cleaned strings
dang_counts = []         # will hold counts, same order as filtered_messages

for message in messages:
    words = message.split()          # break this one message into words
    good_words = []                  # reset PER message -- important!
    dang_count = 0                   # reset PER message -- important!

    for word in words:               # inner loop: walk every word in THIS message
        if word == "dang":
            dang_count += 1          # don't keep it, just count it
        else:
            good_words.append(word)  # keep it

    filtered_message = " ".join(good_words)  # rebuild the sentence from surviving words
    filtered_messages.append(filtered_message)  # append happens together --
    dang_counts.append(dang_count)              # -- so indexes stay in sync

print(filtered_messages)
print(dang_counts)
````

## 4. Putting it in a function

Same logic, wrapped so it can be reused on any list of messages, and returns both lists together.

````python
def filter_messages(messages):
    filtered_messages = []
    dang_counts = []

    for message in messages:
        words = message.split()
        good_words = []
        dang_count = 0

        for word in words:
            if word == "dang":
                dang_count += 1
            else:
                good_words.append(word)

        filtered_message = " ".join(good_words)
        filtered_messages.append(filtered_message)
        dang_counts.append(dang_count)

    return filtered_messages, dang_counts
````

````python
def filter_messages(messages):
    filtered_messages = []   # will hold the cleaned message strings
    dang_counts = []         # will hold the "dang" count per message, same order/index as above

    for message in messages:                # outer loop: one pass per message in the list
        words = message.split()             # break THIS message into a list of individual words
        good_words = []                     # reset per message -- holds words we're keeping
        dang_count = 0                       # reset per message -- tally of "dang"s found here

        for word in words:                  # inner loop: check every word in THIS message
            if word == "dang":
                dang_count += 1              # don't keep it, just add to the tally
            else:
                good_words.append(word)      # keep it -- add to the surviving words list

        filtered_message = " ".join(good_words)         # rebuild a sentence from surviving words
        filtered_messages.append(filtered_message)       # store this message's cleaned result
        dang_counts.append(dang_count)                    # store this message's count, same index

    return filtered_messages, dang_counts   # hand back both parallel lists together
````

````python
#ALT versoin
def filter_messages(messages):
    filtered_messages = []
    dang_counts = []

    for message in messages:                          # loop over each message first
        words = message.split()                        # break THIS message into words
        good_words = [w for w in words if w != 'dang']  # keep everything except "dang"
        dang_count = words.count('dang')                # count how many "dang"s were in THIS message
        filtered_message = ' '.join(good_words)          # rebuild the cleaned sentence
        filtered_messages.append(filtered_message)        # names now match consistently
        dang_counts.append(dang_count)

    return filtered_messages, dang_counts
````

## 6. Common beginner mistakes

- **Punctuation attached to the word breaks the match.** `"dang!"` is not equal to `"dang"` -- `==` compares the whole string, punctuation included. Try the cell below and see.
- **Forgetting to reset `good_words` / `dang_count` inside the outer loop.** If they're created *before* the outer loop instead of inside it, words from message 1 leak into message 2.
- **Appending to the two result lists inside the inner loop instead of after it** — that appends once per *word* instead of once per *message*, breaking the parallel-index structure.

````python
# Demonstrating the punctuation edge case
edge_case = ["get that dang! thing out of here"]
print(filter_messages(edge_case))
# "dang!" (with the exclamation point) does NOT match "dang" exactly,
# so it will NOT be removed or counted -- this is a real limitation of == on strings
````

## 7. Real-world connection

This exact pattern -- loop over records, tokenize text, filter/count based on a condition, build parallel result lists -- is the backbone of basic text data cleaning in data analysis: removing stopwords, counting keyword mentions per review, flagging banned terms in user-generated content, etc. The `.split()` / loop / `.join()` combination you just used is a simplified version of what real text-preprocessing pipelines (like in `pandas` or `nltk`) do internally.

## 8. Alternative approach (more compact, tradeoff)

A list comprehension plus `.count()` can replace the inner loop:

```python
good_words = [w for w in words if w != "dang"]
dang_count = words.count("dang")
```

This is shorter and considered more "Pythonic," but it hides the explicit counting logic. Worth knowing it exists -- but get comfortable with the explicit loop version first, since that's the mechanic you're actively building right now.
