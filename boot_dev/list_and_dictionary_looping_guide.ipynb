# Looping Through Lists and Dictionaries

This guide explains why these two loops look similar:

```python
for room, area in house:
```

and

```python
for key, value in dict1.items():
```

They both use **unpacking**, but they iterate over different data structures.


## 1. The main concept: unpacking

When Python receives a two-item object during a loop, it can place each item into a separate variable.

General pattern:

```python
for first_value, second_value in iterable:
    print(first_value, second_value)
```

The iterable must produce exactly two values on each iteration.


````python
# A list of lists.
# Each inner list contains exactly two values:
# 1. the room name
# 2. the room area

house = [
    ["hallway", 11.25],
    ["kitchen", 18.0],
    ["living room", 20.0],
    ["bedroom", 10.75],
    ["bathroom", 9.50]
]

# On each iteration, Python takes one inner list.
# Example first iteration:
# ["hallway", 11.25]
#
# Python unpacks it as:
# room = "hallway"
# area = 11.25

for room, area in house:
    print(f"The {room} is {area} sqm")
````

### What happens during the first iteration?

Python takes this inner list:

```python
["hallway", 11.25]
```

and assigns:

```python
room = "hallway"
area = 11.25
```

This is called **sequence unpacking**.


````python
# You can see the same idea without a loop.

room_data = ["hallway", 11.25]

# Unpack the two values into two variables.
room, area = room_data

print(room)
print(area)
````

## 2. Looping through a dictionary

A dictionary stores **key-value pairs**.

Example:

```python
dict1 = {
    "health": 100,
    "mana": 50
}
```

Here:

- `"health"` is a key
- `100` is its value
- `"mana"` is a key
- `50` is its value


````python
dict1 = {
    "health": 100,
    "mana": 50,
    "strength": 20
}

# .items() produces key-value pairs.
# Each pair behaves like a two-item tuple.
#
# First iteration:
# ("health", 100)
#
# Python unpacks it as:
# key = "health"
# value = 100

for key, value in dict1.items():
    print(f"{key} has the value {value}")
````

## 3. Are the two loops the same?

They use the same **unpacking pattern**, but they are not looping through the same type of object.

### List of lists

```python
for room, area in house:
```

- `house` is a list
- each item inside `house` is another two-item list
- Python unpacks each inner list

### Dictionary

```python
for key, value in dict1.items():
```

- `dict1` is a dictionary
- `.items()` produces each key-value pair
- Python unpacks each pair

The variable names do not control the behavior. These would also work:

```python
for x, y in house:
    print(x, y)

for x, y in dict1.items():
    print(x, y)
```

Good variable names simply make the code easier to understand.


````python
# Compare the actual objects produced by each loop.

print("Items produced by the list:")
for item in house:
    print(item, type(item))

print("\nItems produced by dictionary.items():")
for item in dict1.items():
    print(item, type(item))
````

Notice the difference:

- The list produces inner **lists**
- `dict1.items()` produces key-value **tuples**

Both contain two values, so both can be unpacked into two variables.


## 4. What happens without `.items()`?

Looping directly through a dictionary returns only its keys.


````python
dict1 = {
    "health": 100,
    "mana": 50
}

# This loops through only the dictionary keys.
for key in dict1:
    print(key)
````

````python
# To retrieve the value, use the key to access the dictionary.

for key in dict1:
    value = dict1[key]
    print(key, value)
````

These two dictionary loops produce the same result:

```python
for key in dict1:
    value = dict1[key]
```

and:

```python
for key, value in dict1.items():
```

The `.items()` version is more direct when both the key and value are needed.


## 5. Applying this to the merge function

The goal is to copy every key-value pair from two dictionaries into one new dictionary.


````python
def merge(dict1, dict2):
    # Create one empty dictionary to hold the final result.
    merged_dict = {}

    # .items() gives us each key and its matching value.
    for key, value in dict1.items():
        # Add the key-value pair to the new dictionary.
        merged_dict[key] = value

    # Repeat the process for the second dictionary.
    for key, value in dict2.items():
        # If this key already exists, dict2's value replaces dict1's value.
        merged_dict[key] = value

    # Send the finished dictionary back to the caller.
    return merged_dict


first = {
    "health": 100,
    "mana": 50
}

second = {
    "strength": 20,
    "speed": 15
}

result = merge(first, second)
print(result)
````

## 6. Why `append()` does not work here

`append()` belongs to lists:

```python
my_list.append("new item")
```

Dictionaries use key assignment:

```python
my_dictionary[key] = value
```

So this is correct:

```python
merged_dict[key] = value
```

This is not correct for a dictionary:

```python
merged_dict.append(value)
```


````python
# List operation
items = []
items.append("potion")
print(items)

# Dictionary operation
inventory = {}
inventory["potion"] = 3
print(inventory)
````

## 7. Common error: too many or too few values to unpack

This loop expects every item to contain exactly two values:

```python
for room, area in house:
```

If one inner list contains three values, Python cannot place three values into only two variables.


````python
# Uncomment this example to see the error.

# broken_house = [
#     ["hallway", 11.25, "downstairs"]
# ]
#
# for room, area in broken_house:
#     print(room, area)

# Error:
# ValueError: too many values to unpack
````

## Final rule to remember

```python
for room, area in house:
```

means:

> Each item in `house` contains two values. Unpack them into `room` and `area`.

```python
for key, value in dict1.items():
```

means:

> Each item produced by `.items()` contains a dictionary key and its value. Unpack them into `key` and `value`.

The shared concept is **unpacking**. The difference is the source of the two values.
