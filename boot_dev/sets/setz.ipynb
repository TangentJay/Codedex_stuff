```python
def remove_duplicates(spells):
    # Tracks spells we have already encountered.
    spell_set = set()

    # Stores the unique spells in their original order.
    spell_list = []

    # Loop through each spell in the original list.
    for spell in spells:

        # Only add the spell if we have not seen it before.
        if spell not in spell_set:
            spell_set.add(spell)
            spell_list.append(spell)

    return spell_list
```
