```python
def reverse_list(items):
    result = []  # new empty list to build the reversed version into
    # start at the last valid index (len(items) - 1),
    # stop before index -1 (i.e., stop once we've included index 0),
    # step backward by 1 each time
    for i in range(len(items) - 1, -1, -1):
        result.append(items[i])  # grab item at position i, add it to the end of result
    return result  # hand back the fully built reversed list
```
